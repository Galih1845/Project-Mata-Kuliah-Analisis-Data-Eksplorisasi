## **Imputasi:** Dalam data ‘weather’, beberapa observasi pada variabel pressure hilang. Lakukan imputasi dengan mengganti nilai yang hilang ini dengan nilai rata-rata pressure pada bulan yang bersangkutan. Bandingkan hasil rata-rata pressure untuk setiap bulannya sebelum dan sesudah melakukan imputasi

```python
#mencari rata rata kolom pressure per bulan
rata_rata_perbulan = df5.groupby('month')['pressure'].mean()
print(rata_rata_perbulan)
```

```python
#mencari missing value kolom pressure
missing_values = df5['pressure'].isnull().sum()
print(missing_values)
```

```python
#menampilkan lokasi missig value
missing_locations = df5[df5['pressure'].isnull()].index
print("Lokasi missing values (baris):\n", missing_locations)
```

```python
#menandai baris yang memiliki missing value pada kolom pressure
df5['missing_pressure'] = df5['pressure'].isna()
print(df5['missing_pressure'])
```

```python
#Mengisi missing value di kolom'pressure' dengn rata-rata per bulan
df5['pressure'] = df5.groupby ('month')['pressure'].transform(lambda x: x.fillna(x.mean()))
print(df5['pressure'])
```

```python
#Menampilkan hanya baris-baris yang diimputasi
imputed_rows = df5[df5['missing_pressure']]
print(imputed_rows)
```

```python
#Mencari rata-rata 'pressure' setelah imputasi berdasarkan "month"
rata_rata_perbulan_setelah_imputasi = df5.groupby('month')['pressure'].mean()
print(rata_rata_perbulan_setelah_imputasi)
```

```python
#membandingkan rata-rata sebelum dan sesudah
print("Rata-rata 'pressure' sebelum imputasi:")
print(rata_rata_perbulan)
print("\nRata-rata 'pressure' setelah imputasi:")
print(rata_rata_perbulan_setelah_imputasi)
```

Rata-rata kolom 'pressure' untuk setiap bulan sebelum dan sesudah imputasi sama persis. Tidak ada perubahan pada rata-rata bulanan, yang berarti nilai yang hilang diisi dengan nilai rata-rata per bulannya tanpa mengubah nilai rata-rata tersebut. Hal ini menunjukkan bahwa imputasi menggunakan rata-rata bulanan berhasil dilakukan tanpa mengganggu distribusi data yang ada, karena nilai yang hilang di setiap bulan telah diisi dengan rata-rata dari bulan yang bersangkutan.

dioperasikan oleh United, American, atau Delta airlines?**

```python
!pip install gdown
import gdown
# Download the file directly from Google Drive using its ID
file_id = '1eJvhTdJF15YC-Hs0j_b9VLiY9jRr7Klo'
output_file = 'flights.csv'
gdown.download(f'https://drive.google.com/uc?id={file_id}', output_file, quiet=False)
```

```python
#menampilkan data frame 3
import pandas as pd
# Membaca file CSV dan menyimpannya ke dalam DataFrame
df3 = pd.read_csv('flights.csv')

# Menampilkan beberapa baris pertama data
print(df3.head())
```

```python
df3
```

```python
#menghilangkan kolom 'Unnamed; 0'
df3.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
df3
```

```python
#memfilter data df3 untuk menyertakan penerbangan hanya dari maskapai United, American, atau Delta dan tahun 2013
filtered_flights = df3[(df3['carrier'].isin(['UA', 'AA', 'DL'])) & (df3['year'] == 2013)]
```

Di sini, 'UA' mewakili United, 'AA' mewakili American, dan 'DL' mewakili Delta.

```python
# menghitung jumlah penerbangan dalam DataFrame
num_flights = filtered_flights.shape[0]
print(f"Jumlah penerbangan di bandara NYC pada tahun 2013 yang dioperasikan oleh United, American, atau Delta airlines adalah: {num_flights}")
```

Sehingga dapat disimpulkan bahwa Jumlah penerbangan di bandara NYC pada tahun 2013 yang dioperasikan oleh United, American, atau Delta airlines adalah 139.504 penerbangan.
