Lakukan pre-processing pada data yang meliputi mengecek data duplikasi, noise, dan outlier, kemudian jawab pertanyaan dibawah ini

```python
# Silahkan tambahkan beberapa sel sesuai kebutuhan untuk pre-processing data sebelum menjawab pertanyaan-oertanyaan dibawah ini.
```

```python
#1 Berapa persentase customer yang bersedia membayar ('Paying Customer' = yes)?
```

```python
# Jumlah pelanggan yang bersedia membayar ('Yes')
paying_customers = df[df['Paying Customer'] == 'Yes'].shape[0]

# Total pelanggan
total_customers = df.shape[0]

# Persentase pelanggan yang bersedia membayar
persentase_paying_customers = (paying_customers / total_customers) * 100

print(f"Persentase pelanggan yang bersedia membayar: {persentase_paying_customers}%")
```

```python
#2 Berapa persentase customer yang bersedia dikontak kembali ('Do_Not_contact' = no)?
```

```python
# Jumlah pelanggan yang bersedia dikontak kembali ('No')
Do_Not_Contact = df[df['Do_Not_Contact'] == 'No'].shape[0]

# Total pelanggan
total_customers = df.shape[0]

# Persentase pelanggan yang bersedia dikontak kembali
persentase_pelanggan_yang_bersedia_dikontak_kembali = (Do_Not_Contact/ total_customers) * 100

print(f"Persentase pelanggan yang bersedia dikontak kembali: {persentase_pelanggan_yang_bersedia_dikontak_kembali}%")
```

```python
#3 Berapa banyak customer yang memiliki Call Duration di luar batas wajar (outlier), jika menggunakan metode deteksi seperti Z-score?
```

```python
z_score= (df['Call Duration (minutes)']-df['Call Duration (minutes)'].mean())/df['Call Duration (minutes)'].std()
print("banyaknya outlier =", sum(abs(z_score)>3))
```

```python
#4 Berapa banyak customer yang memiliki Call Duration di luar batas wajar (outlier), jika menggunakan metode deteksi seperti IQR?
```

```python
#Mendeteksi outlier dengan IQR
Q1 = df['Call Duration (minutes)'].quantile(0.25)
Q2 = df['Call Duration (minutes)'].quantile(0.50)
Q3 = df['Call Duration (minutes)'].quantile(0.75)

IQR = Q3 - Q1 #IQR is interquatile range
print("Q1={}, Q3={}, IQR={}".format(Q1, Q3, IQR))

import seaborn as sns
p = sns.boxplot(x="Call Duration (minutes)", data=df)
```

Dari hasil diatas, terdapat 3 customer yang memiliki call duration diatas batas wajar (outlier)

```python
#5 Bandingkan hasil mean dan standard deviasi Call Duration sebelum penanganan outlier dan sesudah penanganan outlier menggunakan metode capping dan trimming?
```

```python
#Trimming - buang data outlier
outlier = abs(z_score) > 3
df_baru=df[-outlier]
print("Data Sebelum outlier dibuang:", df.shape)
print("Data Setelah outlier dibuang:", df_baru.shape)
```

```python
p= sns.histplot(df_baru['Call Duration (minutes)'])
```

```python
df['Call Duration (minutes)'].describe()
```

```python
df_baru['Call Duration (minutes)'].describe()
```

```python
#Capping - ganti nilai outlier dengan nilai upper limit atau lower limit
upper_limit = df['Call Duration (minutes)'].mean() +3*df['Call Duration (minutes)'].std()
lower_limit = df['Call Duration (minutes)'].mean() -3*df['Call Duration (minutes)'].std()

print(upper_limit, lower_limit)
```

```python
df_baru = df.copy()
df_baru.loc[df_baru['Call Duration (minutes)'] > upper_limit, 'Call Duration (minutes)'] = upper_limit
df_baru.loc[df_baru['Call Duration (minutes)'] < lower_limit, 'Call Duration (minutes)'] = lower_limit
```
