## 1. Siapakah diantara laki-laki dan perempuan yang memiliki rata-rata harapan hidup paling tinggi? Jelaskan!

```python
df
```

```python
# Pisahkan data untuk Laki-laki dan Perempuan sesuai dengan penulisan yang benar
laki_laki = df.xs('Laki-laki', level='Jenis Kelamin')  # Gunakan 'Laki-laki' sesuai dengan indeks
perempuan = df.xs('Perempuan', level='Jenis Kelamin')

# Hitung rata-rata untuk masing-masing jenis kelamin
rata_rata_laki_laki = laki_laki['Angka Harapan Hidup (AHH)'].mean()
rata_rata_perempuan = perempuan['Angka Harapan Hidup (AHH)'].mean()

# Cetak hasil
print(f"Rata-rata Angka Harapan Hidup Laki-laki: {rata_rata_laki_laki}")
print(f"Rata-rata Angka Harapan Hidup Perempuan: {rata_rata_perempuan}")
```

Jadi berdasarkan hasil yang didapat kita mengetahui bahwa perempuan memiliki rata-rata harapan hidup lebih besar daripada laki-laki yaitu :
Perempuan sebesar 72.25068571428571 sedangkan laki-laki sebesar 68.3928

## 2. Provinsi manakah yang memiliki rata-rata harapan hidup paling tinggi? Dan provinsi manakah dengan rata-rata harapan hidup paling rendah?

```python
#hitung rata-rata untuk masing masing provinsi
rata_rata_provinsi = df.groupby('Provinsi')['Angka Harapan Hidup (AHH)'].mean()
#cetak hasil
rata_rata_provinsi
```

```python
#mengurutkan dari yang terbesar
rata_rata_provinsi.sort_values(ascending=False)
```

```python
#menampilkan rata-rata provinsi yang paling besar berdasarkan nama provinsi nya
rata_rata_provinsi.idxmax()
```

```python
rata_rata_provinsi.max()
```

```python
#menampilkan rata-rata provinsi yang paling kecil berdasarkan nama provinsi nya
rata_rata_provinsi.idxmin()
```

```python
rata_rata_provinsi.min()
```

Dapat disimpulkan provinsi yang memiliki rata-rata harapan hidup terbesar yaitu Provinsi DI Yogyakarta dengan 75,074 dan rata-rata harapan hidup terkecil yaitu Provinsi Sulawesi Barat dengan 65,399

## 3. Bagaimana trend rata-rata harapan hidup perempuan dan laki-laki di sepanjang tahun? Apakah naik atau turun atau berfluktuasi? Jelaskan!

```python
df
```

```python
#cek apakah ada outlier
z_score= (df['Angka Harapan Hidup (AHH)']-df['Angka Harapan Hidup (AHH)'].mean())/df['Angka Harapan Hidup (AHH)'].std()
print("banyaknya outlier =", sum(abs(z_score)>3))
```

```python
import matplotlib.pyplot as plt
# Hitung rata-rata harapan hidup berdasarkan tahun dan jenis kelamin
avg_life_expectancy = df.groupby(['Tahun', 'Jenis Kelamin'])['Angka Harapan Hidup (AHH)'].mean().unstack()

# Tampilkan hasil perhitungan rata-rata
avg_life_expectancy
```

```python
# Buat grafik manual dengan matplotlib
plt.figure(figsize=(10, 6))

# Plot garis untuk Laki-laki
plt.plot(avg_life_expectancy.index, avg_life_expectancy['Laki-laki'], label='Laki-laki', color='blue', marker='o')

# Plot garis untuk Perempuan
plt.plot(avg_life_expectancy.index, avg_life_expectancy['Perempuan'], label='Perempuan', color='red', marker='o')

# Tambahkan judul dan label sumbu
plt.title('Trend Rata-rata Harapan Hidup Laki-laki dan Perempuan di Indonesia')
plt.xlabel('Tahun')
plt.ylabel('Rata-rata Angka Harapan Hidup')

# Tampilkan legenda
plt.legend()

# Tambahkan grid
plt.grid(True)

# Tampilkan grafik
plt.show()
```

Dapat disimpulkan bahwa trend rata-rata harapan hidup baik perempuan maupun laki-laki adalah naik. Hal ini menunjukkan adanya peningkatan kualitas hidup masyarakat secara keseluruhan.Kenaikan ini mencerminkan adanya perbaikan dalam akses layanan kesehatan, pengendalian penyakit, serta kondisi sosial dan ekonomi yang lebih baik.
